## CELLULE 5 : Effectifs des esp√®ces

```python
print("\n" + "=" * 80)
print(" " * 30 + "EXERCICE 1")
print(" " * 22 + "DISTRIBUTION DES ESP√àCES")
print("=" * 80)

# 1. Calculer les effectifs
effectifs = df['Species'].value_counts()
pourcentages = df['Species'].value_counts(normalize=True) * 100

print("\nüìä EFFECTIFS PAR ESP√àCE :")
print("-" * 40)
for espece, effectif in effectifs.items():
    pct = pourcentages[espece]
    print(f"  {espece.capitalize():12} : {effectif:3d} √©chantillons ({pct:.1f}%)")
```

## CELLULE 6 : Visualisations - Diff√©rents types de graphiques

```python
# 2. Cr√©er les diff√©rents types de graphiques
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('EXERCICE 1 - Distribution des Esp√®ces d\'Iris\nComparaison de diff√©rentes repr√©sentations', 
             fontsize=16, fontweight='bold', y=1.00)

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']

# 2a. Histogramme (Barres verticales)
axes[0, 0].bar(effectifs.index, effectifs.values, color=colors, edgecolor='black', linewidth=1.5)
axes[0, 0].set_title('A) Histogramme (Barres verticales)', fontweight='bold')
axes[0, 0].set_xlabel('Esp√®ce')
axes[0, 0].set_ylabel('Effectif')
axes[0, 0].grid(axis='y', alpha=0.3)
for i, (espece, effectif) in enumerate(effectifs.items()):
    axes[0, 0].text(i, effectif + 1, str(effectif), ha='center', fontweight='bold')

# 2b. Diagramme en secteurs (Pie chart)
axes[0, 1].pie(effectifs.values, labels=effectifs.index, autopct='%1.1f%%',
               colors=colors, startangle=90, explode=(0.05, 0.05, 0.05),
               shadow=True, textprops={'fontweight': 'bold'})
axes[0, 1].set_title('B) Diagramme en secteurs (Camembert)', fontweight='bold')

# 2c. Barres horizontales
axes[0, 2].barh(effectifs.index, effectifs.values, color=colors, edgecolor='black', linewidth=1.5)
axes[0, 2].set_title('C) Barres horizontales', fontweight='bold')
axes[0, 2].set_xlabel('Effectif')
axes[0, 2].set_ylabel('Esp√®ce')
axes[0, 2].grid(axis='x', alpha=0.3)
for i, (espece, effectif) in enumerate(effectifs.items()):
    axes[0, 2].text(effectif + 1, i, str(effectif), va='center', fontweight='bold')

# 2d. Diagramme en barres group√©es avec Seaborn
sns.countplot(data=df, x='Species', ax=axes[1, 0], palette=colors, edgecolor='black', linewidth=1.5)
axes[1, 0].set_title('D) Barres group√©es (Seaborn)', fontweight='bold')
axes[1, 0].set_xlabel('Esp√®ce')
axes[1, 0].set_ylabel('Effectif')
axes[1, 0].grid(axis='y', alpha=0.3)

# 2e. Diagramme en cascade (Waterfall - simul√©)
cumulative = effectifs.cumsum()
axes[1, 1].bar(range(len(effectifs)), effectifs.values, color=colors, edgecolor='black', linewidth=1.5)
axes[1, 1].set_title('E) Diagramme en cascade', fontweight='bold')
axes[1, 1].set_xticks(range(len(effectifs)))
axes[1, 1].set_xticklabels(effectifs.index)
axes[1, 1].set_xlabel('Esp√®ce')
axes[1, 1].set_ylabel('Effectif')
axes[1, 1].grid(axis='y', alpha=0.3)
# Ajouter les lignes de cumul
for i in range(len(cumulative)-1):
    axes[1, 1].plot([i, i+1], [cumulative.iloc[i], cumulative.iloc[i]], 'k--', alpha=0.5)

# 2f. Diagramme en anneau (Donut)
wedges, texts, autotexts = axes[1, 2].pie(effectifs.values, labels=effectifs.index, 
                                           autopct='%1.1f%%', colors=colors,
                                           startangle=90, pctdistance=0.85,
                                           textprops={'fontweight': 'bold'})
centre_circle = plt.Circle((0,0), 0.70, fc='white')
axes[1, 2].add_artist(centre_circle)
axes[1, 2].set_title('F) Diagramme en anneau (Donut)', fontweight='bold')

plt.tight_layout()
plt.savefig('../visualizations/exercice1_distribution_especes.png', dpi=300, bbox_inches='tight')
plt.show()
```

## CELLULE 8 : Analyse d√©taill√©e de petal_length

```python
print("\n" + "=" * 80)
print(" " * 30 + "EXERCICE 2")
print(" " * 18 + "VARIABLES QUANTITATIVES")
print("=" * 80)

# 1. Analyse d√©taill√©e de petal_length
print("\nüìè ANALYSE D√âTAILL√âE : Longueur du p√©tale (petal_length)")
print("-" * 80)

petal_length = df['PetalLength']

# Statistiques de position
print(f"\nüìä Statistiques de position :")
print(f"   Moyenne      : {petal_length.mean():.3f} cm")
print(f"   M√©diane      : {petal_length.median():.3f} cm")
print(f"   Mode         : {petal_length.mode()[0]:.3f} cm")

# Statistiques de dispersion
print(f"\nüìà Statistiques de dispersion :")
print(f"   √âcart-type   : {petal_length.std():.3f} cm")
print(f"   Variance     : {petal_length.var():.3f} cm¬≤")
print(f"   √âtendue      : {petal_length.max() - petal_length.min():.3f} cm")
print(f"   IQR (Q3-Q1)  : {petal_length.quantile(0.75) - petal_length.quantile(0.25):.3f} cm")

# Valeurs extr√™mes
print(f"\nüî¢ Valeurs extr√™mes :")
print(f"   Minimum      : {petal_length.min():.3f} cm")
print(f"   Maximum      : {petal_length.max():.3f} cm")
print(f"   Q1 (25%)     : {petal_length.quantile(0.25):.3f} cm")
print(f"   Q2 (50%)     : {petal_length.quantile(0.50):.3f} cm")
print(f"   Q3 (75%)     : {petal_length.quantile(0.75):.3f} cm")

# Forme de la distribution
print(f"\nüìâ Forme de la distribution :")
print(f"   Asym√©trie    : {petal_length.skew():.3f} {'(asym√©trique √† droite)' if petal_length.skew() > 0 else '(asym√©trique √† gauche)'}")
print(f"   Aplatissement: {petal_length.kurtosis():.3f} {'(distribution pointue)' if petal_length.kurtosis() > 0 else '(distribution aplatie)'}")
```

## CELLULE 9 : Histogrammes de toutes les variables

```python
# 2. Visualisation de toutes les variables quantitatives
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('EXERCICE 2 - Distribution des Variables Quantitatives', 
             fontsize=16, fontweight='bold')

variables = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
titles = ['Longueur du s√©pale', 'Largeur du s√©pale', 
          'Longueur du p√©tale', 'Largeur du p√©tale']
colors_var = ['#E74C3C', '#3498DB', '#2ECC71', '#F39C12']

for idx, (var, title, color) in enumerate(zip(variables, titles, colors_var)):
    ax = axes[idx // 2, idx % 2]
    
    # Histogramme avec courbe de densit√©
    n, bins, patches = ax.hist(df[var], bins=20, color=color, alpha=0.7, edgecolor='black', linewidth=1.2)
    
    # Lignes de r√©f√©rence
    mean_val = df[var].mean()
    median_val = df[var].median()
    
    ax.axvline(mean_val, color='red', linestyle='--', linewidth=2.5, 
               label=f'Moyenne: {mean_val:.2f}')
    ax.axvline(median_val, color='blue', linestyle='--', linewidth=2.5,
               label=f'M√©diane: {median_val:.2f}')
    
    # Labels et titre
    ax.set_title(f'{title}', fontsize=14, fontweight='bold')
    ax.set_xlabel(f'{title} (cm)', fontsize=11)
    ax.set_ylabel('Fr√©quence', fontsize=11)
    ax.legend(loc='upper right')
    ax.grid(alpha=0.3)
    
    # Ajouter les statistiques
    stats_text = f'Œº = {mean_val:.2f}\nœÉ = {df[var].std():.2f}\nMin = {df[var].min():.2f}\nMax = {df[var].max():.2f}'
    ax.text(0.02, 0.98, stats_text, transform=ax.transAxes, 
            fontsize=9, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.savefig('../visualizations/exercice2_variables_quantitatives.png', dpi=300, bbox_inches='tight')
plt.show()
```

## CELLULE 10 : Tableau r√©capitulatif et conclusion

```python
# 3. Tableau r√©capitulatif de toutes les variables
print("\nüìä TABLEAU R√âCAPITULATIF - TOUTES LES VARIABLES")
print("=" * 80)

summary_stats = pd.DataFrame({
    'Variable': titles,
    'Moyenne': [df[var].mean() for var in variables],
    'M√©diane': [df[var].median() for var in variables],
    '√âcart-type': [df[var].std() for var in variables],
    'Min': [df[var].min() for var in variables],
    'Max': [df[var].max() for var in variables],
    '√âtendue': [df[var].max() - df[var].min() for var in variables]
})

print(summary_stats.to_string(index=False))

print("\n‚úÖ CONCLUSION EXERCICE 2 :")
print("-" * 80)
print("""
Observations importantes :
1. Les dimensions des p√©tales (longueur et largeur) ont une plus grande variabilit√©
2. Les dimensions des s√©pales sont plus homog√®nes
3. Toutes les distributions sont relativement sym√©triques
4. Il y a une s√©paration nette entre les esp√®ces (√† confirmer avec l'exercice 4)
""")
```

## CELLULE 16 : Graphique radar

```python
# 3. Graphique radar pour comparer les esp√®ces
from math import pi

categories = ['Longueur\ns√©pale', 'Largeur\ns√©pale', 'Longueur\np√©tale', 'Largeur\np√©tale']
N = len(categories)

# Calculer les moyennes par esp√®ce
species_means = df.groupby('Species')[variables].mean()

# Cr√©er le graphique radar
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(polar=True))

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']
for (species, color) in zip(species_means.index, colors):
    values = species_means.loc[species].values.flatten().tolist()
    values += values[:1]
    ax.plot(angles, values, 'o-', linewidth=2, label=species, color=color)
    ax.fill(angles, values, alpha=0.25, color=color)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories, size=12)
ax.set_ylim(0, 7)
ax.set_title('Graphique Radar - Moyennes par esp√®ce', 
             size=16, fontweight='bold', pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
ax.grid(True)

plt.savefig('../visualizations/exercice5_radar_chart.png', dpi=300, bbox_inches='tight')
plt.show()
```
